```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])

df.info()
```

```python
df['Order Date']
```

```python
# Feature Engineering
df['days_to_ship'] = (df['Ship Date'] - df['Order Date']).dt.days
df['Year'] = df['Order Date'].dt.year
df['Profit Margin (%)'] = (df['Profit'] / df['Sales']) * 100
df
```

```python
#Aggregate total Sales and Profit by: Region, Category, Sub-Category
agg_df = df.groupby(['Region', 'Category', 'Sub-Category'])[['Sales', 'Profit']].sum().reset_index()
agg_df
```

```python
#Identify top 5 and bottom 5 sub-categories by Profit

subcat_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)
top_5 = subcat_profit.head(5)
bottom_5 = subcat_profit.tail(5)
top_5
```

#Yearly sales trend line

```python
yearly_sales = df.groupby('Year')['Sales'].sum()

plt.figure(figsize=(8, 4))
yearly_sales.plot(kind='line', marker='o', color='green')
plt.title('Year-over-year total sales')
plt.ylabel('Sales ($)')
plt.xticks(yearly_sales.index)
plt.grid(True)
plt.show()
```
